## Running a Random Forest

### Import required libraries

```python
import pandas as pd
import sklearn.metrics
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

### Import and estimate data

```python
data = pd.read_csv("addhealth.csv")
data_clean = data.dropna()
data_clean.dtypes
data_clean.describe()
```

### Data preproccessing

```python
features = list(data_clean.columns.values)
X = data_clean[features]
X.drop('TREG1', axis=1, inplace=True)
y = data_clean[['TREG1']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

###  Fitting random forest classifier

```python
RF_clf = RandomForestClassifier(n_estimators=25, random_state=777)
RF_clf = RF_clf.fit(X_train, y_train)
predictions = RF_clf.predict(X_test)
```

### Validate results

```python
print("Confusion matrix:\n", sklearn.metrics.confusion_matrix(y_test,predictions))
print("Accuracy: ",sklearn.metrics.accuracy_score(y_test, predictions))
```

### Estimating importance of features

```python
important_features = pd.Series(data=RF_clf.feature_importances_,index=X.columns)
important_features.sort_values(ascending=False,inplace=True)

print(important_features)
```
