## Decision Trees

### Import required libraries

```python
import pandas as pd
import sklearn.metrics
from numpy.lib.format import magic
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from io import StringIO
from IPython.display import Image
import pydotplus
RND_STATE = 55324
```

### Import and estimate data

```python
data = pd.read_csv("addhealth.csv")
data_clean = data.dropna()
data_clean.dtypes
data_clean.describe()
```

### Choose predictors and split data to train and test

```python
features = ['BIO_SEX','GPA1', 'ALCEVR1', 'WHITE', 'BLACK']
X = data_clean[features]
y = data_clean[['TREG1']]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

###  Fitting decision tree

```python
DT_clf = DecisionTreeClassifier(min_samples_split=20, random_state=99)
DT_clf = DT_clf.fit(X_train, y_train)
predictions = DT_clf.predict(X_test)
```

### Validate results

```python
print("Confusion matrix:\n", sklearn.metrics.confusion_matrix(y_test,predictions))
print("Accuracy: ",sklearn.metrics.accuracy_score(y_test, predictions))
```

### Plot 

```python
out = StringIO()
tree.export_graphviz(DT_clf, out_file=out, feature_names=["sex", "gpa", "alcohol", "white", "black"],proportion=True, filled=True, max_depth=4)
graph=pydotplus.graph_from_dot_data(out.getvalue())
img = Image(data=graph.create_png())
img 
```
